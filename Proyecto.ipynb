El tipo de cambio afecta las exportaciones de México de manera significativa, ya que influye en la competitividad de los productos naciones en el mercado internacional.

Un peso débil puede hacer que los productos mexicanos sean más competitivos y aumentar las exportaciones.

Un peso fuerte puede encarecer los productos mexicanos y reducir la demanda externa.

La evidencia empírica muestra que la devaluación del tipo de cambio tiene un efecto positivo de crecimiento, dado que aumentan las exportaciones y fortalece el sector de bienes comerciables o industrial en la economía.

Acontinuación se muestra la base de datos del valor de las exportaciones anuales de 2007 a 2024 de México, así como de sus respectivas entidades federativas, extraida de la pagina oficial del INEGI.

**Filas:** representan diferentes entidades federeativas de México y un total nacional.

**Columnas:** Representan diferentes años (2007-2024).

```python
import pandas as pd
df_exp=pd.read_excel("/content/drive/MyDrive/EAEF_Entidad.xlsx")
df_exp
```

Vamos a obtener algunas estadisticas descriptivas con los valores proporcinados en la base de datos.

Como primer analisis, veremos una grafica de barras de las exportaciones anuales a nivel nacional y ontendremos datos por medio de la grafica.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker
```

```python
fila_nacional = df_exp.iloc[0,1:]
años = fila_nacional.index.astype(str).astype(int)
valores = fila_nacional.values
plt.figure(figsize=(12,6))
plt.bar(años, valores, color='skyblue', edgecolor='black')
plt.xlabel("Años")
plt.ylabel("Valor (Dolares)")
plt.title("Exportaciones Mexicanas")
plt.xticks(años, rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: f'{int(x):,}'))
plt.show()
```

**Observaciones:**

La grafica de barras muestra la tendencia al alza de las exportaciones mexicanas, sin embargo el crecimiento no es uniforme, algunos años muestran incrementos más pronunciados que otros, hay ciertas fluctuaciones en algunos años como lo son 2009, 2014, 2016 y 2020.

Tambien se observa una aceleración en el crecimiento de las exportaciones en los ultimos años (2021 a 2024) mostrando un incremento más rapido que en años anteriores



```python
plt.figure(figsize=(8,6))
plt.boxplot(fila_nacional, vert=False, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.xlabel("Valor (Dolares)")
plt.title("Exportaciones Mexicanas")
plt.xticks(rotation=45)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.gca().xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: f'{int(x):,}'))
plt.show()
```

**Observaciones:**

la mediana de se enuentra aproximadamente en los 375,000,000 de dólares. Esto significa que la mitad de las exportaciones tienen un valor inferior a esta cantidad y la otra mitad superior.

El 25% de las observaciones en la grafica se ubican alrededor de los 300,000,000 de dólares.

El 75% de las observaciones se ubican alrededor de los 450,000,000 dólares.


Ahora analizaremos el total de exportaciones de cada entidad federativa, creando una nueva columna con el nombre "Total"

```python
df_exp["Total"] = df_exp.iloc[:,1:].sum(axis=1)
df_exp[['Entidad Federativa','Total']]
```

```python
df_estados = df_exp.iloc[1:,:]
max_value = df_estados['Total'].max()
min_value = df_estados['Total'].min()

max_entity = df_exp.loc[df_exp['Total'] == max_value, 'Entidad Federativa'].iloc[0]
min_entity = df_exp.loc[df_exp['Total'] == min_value, 'Entidad Federativa'].iloc[0]

print(f"Máximo valor: {max_value} ({max_entity})")
print(f"Mínimo valor: {min_value} ({min_entity})")
```

El estado con el mayor valor de exportaciones es Chihuahua, Mientras el estado con menor valor de exportaciones es Quintana Roo

```python
df_estados = df_exp.iloc[1:,:]
df_estados = df_estados.sort_values(by="Total", ascending=False)
plt.figure(figsize=(12,8))
plt.barh(df_estados.iloc[:,0],df_estados["Total"],color="cornflowerblue")
plt.xlabel("total de exportaciones (Dólares)")
plt.ylabel("Entidad Federativa")
plt.title("Total de exportaciones por estado")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.gca().xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: f'{int(x):,}'))
plt.show()
```

```python
top10 = df_estados.sort_values(by="Total", ascending=False).head(10)
plt.figure(figsize=(8,8))
plt.pie(top10["Total"], labels=top10["Entidad Federativa"], autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title("Top 10 Estados por Total de Exportaciones")
plt.axis('equal')
plt.show()
```

Ahora analizaremos la base de datos del tipo de cambio en México, la cúal va del año 1990 al año 2025, con un precio del dolar promedio por año, precio minimo por año al igual que precio máximo por año.

```python
df_camb=pd.read_excel("/content/drive/MyDrive/Cambio_Dolar.xlsx")
df_camb
```

```python
max_val = df_camb['Precio (promedio)'].max()
min_val = df_camb['Precio (promedio)'].min()

max_pre = df_camb.loc[df_camb['Precio (promedio)'] == max_val, 'Año'].iloc[0]
min_pre = df_camb.loc[df_camb['Precio (promedio)'] == min_val, 'Año'].iloc[0]

print(f"Máximo valor: {max_val} ({max_pre})")
print(f"Mínimo valor: {min_val} ({min_pre})")
```

```python
min_val = df_camb['minimo'].min()
max_val = df_camb['maximo'].max()

max_pre = df_camb.loc[df_camb['maximo'] == max_val, 'Año'].iloc[0]
min_pre = df_camb.loc[df_camb['minimo'] == min_val, 'Año'].iloc[0]

print(f"Máximo valor: {max_val} ({max_pre})")
print(f"Mínimo valor: {min_val} ({min_pre})")
```

```python
plt.figure(figsize=(12,6))
# Extract the 'Año' and 'Precio (promedio)' columns from the DataFrame
Año = df_camb['Año']
Precio_promedio = df_camb['Precio (promedio)']  # Rename to a valid variable name
# Plot the bar chart
plt.bar(Año, Precio_promedio, color='skyblue', edgecolor='black')  # Use the extracted data
plt.xlabel("Años")
plt.ylabel("Cambio (pesos)")
plt.title("Valor del tipo de cambio (Dolar a Peso)")
plt.xticks(Año, rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: f'{int(x):,}'))
plt.show()
```

Desde 1990 hasta 1995, el tipo de cambio se mantuvo relativamente bajo y con un crecimiento moderado, a partir de de 1995 se observa un crecimiento más pronunciado y constantemente, con algunas fluctuaciones menores a lo largo de los años.
